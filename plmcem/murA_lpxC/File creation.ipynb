```python
import pandas as pd
import numpy as np
from alignmenttools.MultipleSequenceAlignment import readAlignment
```

```python
df = pd.read_csv('SB_murA_lpxC_scores_081722.csv')
```

```python
msa = readAlignment('murA_lpxC_b08_init.fa')
```

```python
df_l = set(df['full_id'])
msa_l = set(msa.ids)
```

```python
print(len(df_l.intersection(msa_l)))
print(len(df_l))
print(len(msa_l))
```

```python
df_ids = list(df['full_id'])
cluster1_ids = list(df[df['prob'] < 0.5]['full_id'])
cluster2_ids = list(df[df['prob'] > 0.5]['full_id'])
```

```python
full = msa.search_ids(df_ids)
cluster1 = msa.search_ids(cluster1_ids)
cluster2 = msa.search_ids(cluster2_ids)
```

```python
full.save('murA_lpxC_filtered.fa')
cluster1.save('murA_lpxC_true_cluster0.fa')
cluster2.save('murA_lpxC_true_cluster1.fa')
```
