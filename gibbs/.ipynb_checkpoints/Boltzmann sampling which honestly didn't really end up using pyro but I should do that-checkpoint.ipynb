# 1. NUTS simulation of a Normal
Let's try NUTS simulating a Normal distribution through the potential function and see if it actually gives us what we expect.
$$
e^{-(x-\mu)^2/2\sigma} \propto e^{-U(x)}\\
U(x) = \frac{(x-\mu)^2}{2 \sigma^2}
$$
I specifically establish that this works for non-Standard normals and functions under additive constants to the potential function, as it should.  

```python
def give_norm_potential_fn(mu = 0, sigma = 1):
    def potential_fn(in_dict):
        '''
        in_dict: input, dictionary of parameters; has to be this way for HMC to work 
        '''
        #chillin under additive constants
        x = in_dict['x']
        return 0.5 * ((x-mu)/sigma)**2 + 22
    return potential_fn

rng = np.random.default_rng()
mu = rng.standard_normal() * 5
sigma = abs(rng.standard_normal())*3
num_samples = 5000

potential_fn = give_norm_potential_fn(mu, sigma)
nuts_kernel = NUTS(potential_fn = potential_fn)
mcmc = MCMC(nuts_kernel, num_samples = num_samples, warmup_steps = 1000, initial_params = {'x': torch.tensor([1.1])})
mcmc.run()
samples = mcmc.get_samples()
x_samples = torch.flatten(samples['x']).cpu().detach().numpy()
```

```python
fig, ax = plt.subplots(nrows = 2, figsize=(8, 10))
n_bins = 50
n, bins, patches = ax[0].hist(x_samples, n_bins, density=1, histtype='step',
                           cumulative=True, label='Empirical')
y = scipy.stats.norm.pdf(bins, mu, sigma).cumsum()
y /= y[-1]
ax[0].plot(bins, y, 'k--', linewidth=1.5, label='Theoretical')

ax[0].grid(True)
ax[0].legend(loc='right')
ax[0].set_title('CDFs')
ax[0].set_xlabel('x')
ax[0].set_ylabel('Probability')

ax[1].hist(x_samples, n_bins, density = True, label = 'Empirical')
ax[1].plot(bins, scipy.stats.norm.pdf(bins, mu, sigma), label='Theoretical')

ax[1].grid(True)
ax[1].legend(loc='right')
ax[1].set_title('PDFs')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Density')
print('True mu = ' + str(mu) + ', true sigma = '+ str(sigma))
print('Simulated mu = ' + str(np.mean(x_samples)) + ', simulated sigma = ' + str(np.std(x_samples)))
plt.show()
```

HMC is made for continuous so that was a slight waste of time... instead we're going to use implement our own jazz. Base Metropolis-Hastings on a Bernoulli first

```python
rng = np.random.default_rng()
proposal_pmf = 0.7

def proposal_prob(state, next_state):
    if next_state == 0:
        return proposal_pmf
    return 1 - proposal_pmf

def proposal_dist(state):
    proposal_state = int(proposal_pmf < rng.random())
    return proposal_state, proposal_prob(state, proposal_state), proposal_prob(proposal_state, state)

def unnormalized_target_dist(state):
    p = 0.2
    factor = 239
    if state:
        return p * factor
    return (1-p) * factor

def mh_transition(current_state, proposal_dist, unnormalized_target_dist):
    proposal_state, prob, reverse_prob = proposal_dist(current_state)
    acceptance_prob = min(1, unnormalized_target_dist(proposal_state)/unnormalized_target_dist(current_state) * reverse_prob / prob)
    if acceptance_prob >= 1:
        return proposal_state
    if rng.random() < acceptance_prob:
        return proposal_state
    return current_state

state = 0
for i in range(1000):
    state = mh_transition(state, proposal_dist, unnormalized_target_dist)

states = [state]
for i in range(10000):
    states.append(mh_transition(states[-1], proposal_dist, unnormalized_target_dist))
np.mean(states)
```

```python
rng = np.random.default_rng()
proposal_pmf = 0.9

def proposal_prob(state, next_state):
    if next_state == 0:
        return proposal_pmf
    return 1 - proposal_pmf

def proposal_dist(state):
    proposal_state = int(proposal_pmf < rng.random())
    return proposal_state, proposal_prob(state, proposal_state), proposal_prob(proposal_state, state)

def unnormalized_target_dist(state):
    p = 0.2
    factor = 239
    if state:
        return np.log(p) + factor
    return np.log(1-p) + factor

def mh_transition(current_state, proposal_dist, unnormalized_target_dist):
    proposal_state, prob, reverse_prob = proposal_dist(current_state)
    acceptance_prob = min(0, unnormalized_target_dist(proposal_state) - unnormalized_target_dist(current_state) + np.log(reverse_prob/prob))
    if acceptance_prob >= 0:
        return proposal_state
    if np.log(rng.random()) < acceptance_prob:
        return proposal_state
    return current_state

state = 0
for i in range(1000):
    state = mh_transition(state, proposal_dist, unnormalized_target_dist)

states = [state]
for i in range(100000):
    states.append(mh_transition(states[-1], proposal_dist, unnormalized_target_dist))
np.mean(states)
```

# 3. Let's try Metropolis in 2D?
Okay so four states [0, 0], [0, 1], [1, 0], [1, 1]  
$$P([0, 0]) \propto 3\\
P([0, 1]) \propto 1\\
P([1, 0]) \propto 10\\ 
P([1, 1]) \propto 6$$

```python
rng = np.random.default_rng()

def proposal_prob(state, next_state):
    first_eq = (state[0] == next_state[0])
    second_eq = (state[1] == next_state[1])
    if first_eq == second_eq:
        return 0
    return 0.5

def proposal_dist(state):
    proposal_pos = int(rng.random() < 0.5)
    proposal_state = state.copy()
    proposal_state[proposal_pos] = 1 - proposal_state[proposal_pos]
    return proposal_state, proposal_prob(state, proposal_state), proposal_prob(proposal_state, state)

def unnormalized_log_target_dist(state):
    if state == [0, 0]:
        return np.log(3)
    if state == [0, 1]:
        return np.log(1)
    if state == [1, 0]:
        return np.log(10)
    if state == [1,  1]:
        return np.log(6)
    raise Exception('Invalid state ' + str(state))

def mh_transition(current_state, proposal_dist, unnormalized_log_target_dist):
    proposal_state, prob, reverse_prob = proposal_dist(current_state)
    acceptance_prob = min(0, unnormalized_log_target_dist(proposal_state) - unnormalized_log_target_dist(current_state) + np.log(reverse_prob/prob))
    if acceptance_prob >= 0:
        return proposal_state
    if np.log(rng.random()) < acceptance_prob:
        return proposal_state
    return current_state

state = [0, 0]
for i in range(1000):
    state = mh_transition(state, proposal_dist, unnormalized_log_target_dist)

states = [state]
for i in range(100000):
    states.append(mh_transition(states[-1], proposal_dist, unnormalized_log_target_dist))

def pmf(states):
    state_dict = {}
    num_states = len(states)
    for state in states:
        state_dict[tuple(state)] = state_dict.get(tuple(state), 0) + 1
    for state in state_dict:
        state_dict[state] /= num_states
    return state_dict

pmf_dict = pmf(states)
target_dict = {(0, 0): 3/20, (0, 1): 1/20, (1, 0): 10/20, (1, 1): 6/20}
state_list = [(0, 0), (0, 1), (1, 0), (1, 1)]

def print_pmf(target_dict, pmf_dict):
    for state in state_list:
        out_str = str(state) + ': '
        out_str += str(round(pmf_dict[state], 3)) + ', '
        out_str += str(round(target_dict[state], 3))
        print(out_str)
        
print_pmf(target_dict, pmf_dict)
```

# 4. Gibbs - 2D

```python
rng = np.random.default_rng()
alphabet = [0, 1]

def unnormalized_target_dist(state):
    if state == [0, 0]:
        return 3
    if state == [0, 1]:
        return 1
    if state == [1, 0]:
        return 10
    if state == [1, 1]:
        return 6
    raise Exception('Invalid state ' + str(state))

def unnormalized_log_target_dist(state):
    return np.log(unnormalized_target_dist(state))
    
def normalize_dist(x, total = 1):
    if x is list:
        x = np.array(x)
    return x/np.sum(x)*total

def normalize_log_dist(x, total = 0):
    if x is list:
        x = np.array(x)
    return x - (np.sum(x)-total)/np.size(x)

def conditional_dist(state, i, unnormalized_target_dist):
    '''
    state: vector
    i: position being varied (rest are fixed)
    '''
    before = state[:i]
    after = state[(i+1):]
    dist = [unnormalized_target_dist(before + [value] + after) for value in alphabet]
    return normalize_dist(dist)
    
def gibbs_transition_samesweep(state, unnormalized_target_dist):
    current_state = state.copy()
    for i in range(len(state)):
        current_state[i] = np.random.choice(alphabet, p = conditional_dist(current_state, i, unnormalized_target_dist))
    return current_state

def gibbs_transition(state, unnormalized_target_dist):
    current_state = state.copy()
    sweep_order = rng.permutation(range(len(state)))
    for i in sweep_order:
        current_state[i] = np.random.choice(alphabet, p = conditional_dist(current_state, i, unnormalized_target_dist))
    return current_state

state = [0, 0]
for i in range(1000):
#     print(state)
    state = gibbs_transition(state, unnormalized_target_dist)

states = [state]
for i in range(10000):
    states.append(gibbs_transition(states[-1], unnormalized_target_dist))

def pmf(states):
    state_dict = {}
    num_states = len(states)
    for state in states:
        state_dict[tuple(state)] = state_dict.get(tuple(state), 0) + 1
    for state in state_dict:
        state_dict[state] /= num_states
    return state_dict

pmf_dict = pmf(states)
target_dict = {(0, 0): 3/20, (0, 1): 1/20, (1, 0): 10/20, (1, 1): 6/20}
state_list = [(0, 0), (0, 1), (1, 0), (1, 1)]

def print_pmf(target_dict, pmf_dict):
    for state in state_list:
        out_str = str(state) + ': '
        out_str += str(round(pmf_dict[state], 3)) + ', '
        out_str += str(round(target_dict[state], 3))
        print(out_str)
        
print_pmf(target_dict, pmf_dict)
```

```python
rng = np.random.default_rng()
L = 3
alphabet = [0, 1]

def unnormalized_boltzmann(state, hi, Jij):
    if state is tuple:
        state = list(state)
    return np.exp(unnormalized_log_boltzmann(state, hi, Jij))

def unnormalized_log_boltzmann(state, hi, Jij):
    '''
    For calculating the energy which we obviously have better ways to do
    '''
    if state is tuple:
        state = list(state)
    hi_sum = sum(hi[i][val] for i, val in enumerate(state))
    Jij_sum = 0
    for i, vali in enumerate(state):
        for j, valj in enumerate(state[(i+1):]):
            Jij_sum += Jij[i][i+j+1][vali][valj]
    return -1 * (hi_sum + Jij_sum)
    
def normalize_dist(x, total = 1):
    if x is list:
        x = np.array(x)
    return x/np.sum(x)*total

def normalize_log_dist(x, total = 0):
    if x is list:
        x = np.array(x)
    return x - (np.sum(x)-total)/np.size(x)

def conditional_dist(state, i, unnormalized_target_dist, dist_params, alphabet):
    '''
    state: vector
    i: position being varied (rest are fixed)
    '''
    before = state[:i]
    after = state[(i+1):]
    dist = [unnormalized_target_dist(before + [value] + after, *dist_params) for value in alphabet]
#     print(len(dist))
    return normalize_dist(dist)
    
def gibbs_transition_samesweep(state, unnormalized_target_dist, dist_params, alphabet):
    current_state = state.copy()
    for i in range(len(state)):
        current_state[i] = np.random.choice(alphabet, p = conditional_dist(current_state, i, unnormalized_target_dist, dist_params, alphabet))
    return current_state

def gibbs_transition(state, unnormalized_target_dist, dist_params, alphabet):
    current_state = state.copy()
    sweep_order = rng.permutation(range(len(state)))
    for i in sweep_order:
        current_state[i] = np.random.choice(alphabet, p = conditional_dist(current_state, i, unnormalized_target_dist, dist_params, alphabet))
    return current_state

def pmf(states):
    state_dict = {}
    num_states = len(states)
    for state in states:
        state_dict[tuple(state)] = state_dict.get(tuple(state), 0) + 1
    for state in state_dict:
        state_dict[state] /= num_states
    return state_dict

def get_state_list(L, alphabet):
    if L == 0:
        return []
    if L == 1:
        return [(val,) for val in alphabet]
    ret_list = []
    prev = get_state_list(L-1, alphabet)
    for val in alphabet:
        for suff in prev:
            ret_list.append(tuple([val] + list(suff)))
    return ret_list

def print_pmf(pmf_dict, target_dict = None, state_list = None, convert = None):
    if not state_list:
        state_list = set(pmf_dict.keys())
        if target_dict:
            state_list = state_list.union(set(target_dict.keys()))

    for state in state_list:
        if convert == None:
            out_str = str(state) + ': '
        else:
            out_str = str(convert(state)) + ': '
        out_str += str(round(pmf_dict.get(state, 0.), 3)) + ', '
        if target_dict:
            out_str += str(round(target_dict[state], 3))
        print(out_str)    

def run_gibbs(unnormalized_target_dist, dist_params, 
              initial_state = None, 
              L = None, 
              alphabet = None, 
              convert = None, 
              warmup_steps = 1000,
              run_steps = 10000,
              compare_to_true = False,
              return_chain = False
             ):
    if not L:
        L = dist_params[0].shape[0]
    if not alphabet:
        alphabet = range(dist_params[0].shape[1])
    if initial_state:
        state = initial_state
    else:
        state = list([rng.choice(alphabet) for i in range(L)])
    
    for i in range(warmup_steps):
        state = gibbs_transition(state, unnormalized_target_dist, dist_params, alphabet)
    states = [state]
    for i in range(run_steps):
        state = gibbs_transition(states[-1], unnormalized_target_dist, dist_params, alphabet)
        states.append(state)
#         print(state)
#         states.append(gibbs_transition(states[-1], unnormalized_target_dist, dist_params))
        
    pmf_dict = pmf(states)
    if compare_to_true:
        target_dict = {}
        state_list = get_state_list(L, alphabet)
        target_list = normalize_dist([unnormalized_target_dist(state, *dist_params) for state in state_list])
        for i, state in enumerate(state_list):
            target_dict[state] = target_list[i]
        print_pmf(pmf_dict, target_dict, state_list, convert = convert)
    else:
        print_pmf(pmf_dict, convert = convert)
    
    if return_chain:
        return pmf_dict, states
    return states

# dist_params = [hi, Jij]
# state = list(np.random.choice(np.array(alphabet), L))
# for i in range(1000):
# #     print(state)
#     state = gibbs_transition(state, unnormalized_target_dist, dist_params)

# states = [state]
# for i in range(10000):
#     states.append(gibbs_transition(states[-1], unnormalized_target_dist, dist_params))

# pmf_dict = pmf(states)
# target_dict = {}
# state_list = get_state_list(L, alphabet)
# target_list = normalize_dist([unnormalized_target_dist(state, *dist_params) for state in state_list])
# for i, state in enumerate(state_list):
#     target_dict[state] = target_list[i]
        
# print_pmf(target_dict, pmf_dict, state_list)
```
